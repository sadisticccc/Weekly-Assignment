```python
#Q1
def conv(num):
    c = ''
    
    while num != 0:
        r = num / 2
        num = num // 2
        
        if r == num:
            number = '0'
        else:
            number = '1'
            
        c = number + c
        
    if c != '':
        return c
    
    else:
        return '0'
    
dec = int(input("Enter a number in decimal: "))    
print(conv(dec))
```

```python
#Q2
def integer(num):
    factors = []
    
    for i in range(1, num+1):
        if num % i == 0:
            factors.append(i)
    
    return factors

num = int(input("Enter a number: "))
print(integer(num))
```

```python
#Q3
def integer(num):
    factors = []
    
    for i in range(1, num+1):
        if num % i == 0:
            factors.append(i)
    
    return factors

num = int(input("Enter a number: "))
factors = integer(num)

if len(factors) == 2:
    print("It is prime")
    
elif len(factors) > 2:
    print("It is composite")

else:
    print("It is neither prime nor composite.")
```

```python
#Q4
def encrypt(password):
    new = ''
    for i in password:
        if i != ' ':
            new = i + new
    return new

password = input("Enter a password: ")
print(f"Encoded password is {encrypt(password)}")
```

```python
#Q5
import random
def hidden(word):
    sum = ''
    rand_num = random.randint(2, 20)
    
    for i in word:
        rand_char = ''
        
        for j in range(rand_num):
            
            n = random.randint(1,2)
            
            if n == 1:
                rand_char += chr(random.randint(97, 122))
                
            elif n == 2:
                rand_char += chr(random.randint(65, 90))

        if  i != ' ':
            sum = sum + i + rand_char
            
        else:
            sum += ' '
    
    new = sum[:-rand_num]
    
    return new, rand_num
message = input("Enter a message: ")

encrypted, value = hidden(message)

print(encrypted)
```

```python
#Q6
#Run this program only after running the code above (#Q5)

no_space = encrypted.split(" ")

word_list = []

for num in range(len(no_space)):
    word_list.append(no_space[num])

new_word_list = []

for k in word_list:
    new = ''
    for j in range(0, len(k), value+1):
        new += k[j]
    new_word_list.append(new)
    
decrypted = ' '.join(new_word_list)

print(decrypted)
```
